```python
class LibraryItem:
    def __init__(self, title, author, item_id):
        self.title = title
        self.author = author
        self.item_id = item_id
        self.is_available = True

    def borrow_item(self):
        if self.is_available:
            self.is_available = False
            print(f"{self.title} has been borrowed.")
        else:
            print(f"{self.title} is not available.")

    def return_item(self):
        if not self.is_available:
            self.is_available = True
            print(f"{self.title} has been returned.")
        else:
            print(f"{self.title} was not borrowed.")

    def __str__(self):
        availability = "Available" if self.is_available else "Not Available"
        return f"{self.title} by {self.author}, ID: {self.item_id} ({availability})"


class Book(LibraryItem):
    def __init__(self, title, author, item_id, pages):
        super().__init__(title, author, item_id)
        self.pages = pages

    def __str__(self):
        return super().__str__() + f" - Pages: {self.pages}"


class Magazine(LibraryItem):
    def __init__(self, title, author, item_id, issue_date):
        super().__init__(title, author, item_id)
        self.issue_date = issue_date

    def __str__(self):
        return super().__str__() + f" - Issue: {self.issue_date}"


class DVD(LibraryItem):
    def __init__(self, title, author,item_id, duration):
        super().__init__(title, author, item_id)  
        self.duration = duration

    def __str__(self):
        return super().__str__() + f" - Duration: {self.duration} mins"


if __name__ == "__main__":
    book = Book("The Great Gatsby", "F. Scott Fitzgerald", "B101", 200)
    magazine = Magazine("Tech Today", "Various", "M202", "August 2024")
    dvd = DVD("Inception", "Shuhood","D303", 148)

    items = [book, magazine, dvd]

    # Looping through items to demonstrate functionality
    for item in items:
        print(item)          
        item.borrow_item()   
        item.return_item()   
        print()              
```
